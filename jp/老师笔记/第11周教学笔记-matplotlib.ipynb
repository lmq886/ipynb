## 一.pyplot基础语法和常用参数
### 1.pyplot基础语法

第一步 创建画布：plt.figure

第二步 创建子图：plt.subplot

第三步 制作图形：plt.plot、plt.barh、plt.hist、plt.pie、plt.scatter、plt.boxplot……

第六步 显示图形：plt.show

#### 第一步 创建画布
plt.figure语法：
plt.figure(num=None,figsize=None,dpi=None,facecolor=None,edgecolor=None, frameon=True)
#num：图像编号或名称，数字为编号 ，字符串为名称
#figsize：指定figure的宽和高，单位为英寸；
#dpi：参数指定绘图对象的分辨率，即每英寸多少个像素，缺省值为80，1英寸等于2.5cm
#facecolor：背景颜色
#edgecolor：边框颜色
#frameon：是否显示边框

```python
#例2：创建一个空白画布
fig=plt.figure(figsize=(5,5), facecolor = '#FF00FF')
plt.savefig('D:/pyData/例2.png', facecolor=fig.get_facecolor())
plt.show()
```

#### 第二步 创建子图
plt.subplot(nrows,ncols,plot_number)
          #行  #列  #当前子图区
例如：plt.subplot(3,2,4)、（324）   #将画布分为3行2列，取第4格的区域作为当前子图区进行图形的绘制。

3行2列分布图：

     1  2
     3  4
     5  6 

```python
#例3：创建子图
fig=plt.figure(figsize=(5,5), facecolor = 'r')
plt.subplot(3,2,4)
plt.plot([0,2,4,6,8],[3,1,4,5,2])                                   # 第三步 制作图形
plt.xlabel('garde')                                                 # 第四步 美化图形
plt.savefig('D:/pyData/例3.png', facecolor=fig.get_facecolor())     # 第五步 保存图形
plt.show()                                                          # 第六步 显示图形
```

```python
nums = np.arange(0,101)
fig = plt.figure()
#图1
plt.subplot(221)
plt.plot(nums,nums)
plt.title('正比例函数')
plt.xticks((0,20,40,60,80,100))
plt.yticks((0,20,40,60,80,100))
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y=x'])
#图2
plt.subplot(222)
plt.plot(nums,-nums)
plt.title('反比例函数')
plt.xticks((0,20,40,60,80,100))
plt.yticks((-100,-80,-60,-40,-20,0))
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y=-x'])
#图3
plt.subplot(212)
plt.plot(nums,nums**2,'g--')
plt.title('曲线函数')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y=x^2'])
plt.savefig('D:/pyData/例子')
plt.show()
```

```python
#例4：pyplot基础语法综合实例
plt.rcParams['font.family']='SimHei'         #中文正常显示，'SimHei'为黑体，还可以设置'SimSun'为宋体，'KaiTi'为楷体……
plt.rcParams['axes.unicode_minus']=False    #负号正常显示
nums = np.arange(0,101)
fig = plt.figure(figsize=(8,6),dpi=200)
plt.subplot(221)
plt.title('正比例函数')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((0,100))
plt.ylim((0,100))
plt.xticks([0,20,40,60,80,100])
plt.yticks([0,20,40,60,80,100])
plt.plot(nums,nums)
plt.legend(['y=x'])
plt.subplot(222)
plt.title('反比例函数')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((0,100))
plt.ylim((-100,0))
plt.xticks([0,20,40,60,80,100])
plt.yticks([-100,-80,-60,-40,-20,0])
plt.plot(nums,-nums)
plt.legend(['y=-x'])
plt.subplot(212)
plt.title('曲线函数')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(nums,nums**2,'r-.')
plt.legend(['y=x^2'])
plt.savefig(('D:/pyData/例4.png'))
plt.show
```

### 1.绘制折线图
折线图是一种将数据点按照顺序连接起来的图形。比较适合描述和比较多组数据随时间变化的趋势。

plt.plot语法：
plt.plot(x, y, format_string, **kwargs)
#x:  X轴数据，列表或数组
#y:  Y轴数据，列表或数组
#format_string：控制曲线的格式字符串（颜色字符、风格字符、标记字符组成）
#**kwargs：第二组或更多plt.plot(x, y, format_string)

```python
#例5：读取“烧烤店营业额”数据，使用折线图绘制该店全年营业额的变化，要求绿色虚线连接每月数据，并在每月的数据上使用圆点做标记。
data1 = pd.read_excel('D:pyData\烧烤店营业额.xlsx')
plt.figure()
plt.plot(data1['月份'],data1['营业额（万元）'],'g--o')
plt.title('烧烤店2019年营业额趋势图')
plt.xlabel('月份')
plt.ylabel('营业额（万元）')
plt.savefig('D:pyData\例5')
plt.show()
```

```python
#例6：P203 9-1 顾客每多买一件就给优惠1%
basePrice=49   #进价
salePrice=75   #零售价

#定义一个函数，计算购买num个商品时的单价
def compute(num):
    return salePrice*(1-0.01*num)

#生成4个列表
numbers = list(range(1,31))  #用来存储顾客购买数量，题目要求每人最多可以购买30件
earns = []                  #用来存储商场的盈利情况
total = []                   #用来存储顾客消费总额
saves = []                   #用来存储顾客节省的总金额
for num in numbers:
    perPrice = compute(num)                            #购买num个时的成交单价
    earns.append(round(num*(perPrice-basePrice),2))    #（购买单价-进价）*数量=商场盈利，保留2位小数
    total.append(round(num*perPrice,2))                # 购买单价*数量 = 顾客消费总额，保留2位小数
    saves.append(round(num*(salePrice-perPrice),2))    # （零售价-购买单价）*数量 = 节省的总金额，保留2位小数

#绘制图形
plt.plot(numbers,earns)     
plt.plot(numbers,total)
plt.plot(numbers,saves)
plt.xlabel('顾客购买数据（件）')
plt.ylabel('金额（元）')
plt.title('利润最大化')
plt.legend(['商家盈利','顾客总消费','顾客节省'])

#绘制
maxEarn = max(earns)
bestNumber = numbers[earns.index(maxEarn)]
plt.scatter([bestNumber],[maxEarn],marker='*',color='r',s=120)
plt.text(bestNumber-1,maxEarn+200,str(maxEarn),fontsize=15)  
q
plt.show()
```

### 2.绘制散点图
散点图又称散点分布图，是以一个特征为横坐标，另一个特征为纵坐标，利用坐标点的分布形态反映特征间的统计关系。通常适合用于比较垮类别的数据。

plt.scatter语法：
plt.scatter (x, y, s, c, marker, alpha   )
#x,y     表示xy轴对应的数据
#s      表示点的大小
#c      表示点的颜色
#marker  表示点的形状
#alpha   表示点的透明度

```python
#例7：使用散点图绘制随机正态分布小数
x=np.random.normal(0,1,100)         #均值为0，标准差为1的随机正态分布数，一共100个
y=np.random.normal(0,1,100)
plt.scatter(x,y,c='y',marker='v')
plt.show()
```

### 3.绘制饼图
饼图是将各项数据与其总和的比例显示在一张“饼”中，以“饼”的大小来确定每一项的占比，清楚反应部分与部分、部分与整体之间的关系。
plt.pie语法：
plt.pie(x,explode=None,labels=None,autopct=None,startangle=None)
#x：           饼块的各个数据
#explode：     饼块离中心的距离
#labels：      饼块的标签
#autopct：    饼块中百分比数值的设置
#shadow：      饼图的阴影显示
#startangle：  起始绘图的角度

```python
#例9：模拟电商市场占有率
names =['淘宝','京东','唯品会','天猫']
sizes =[1143,398,159,547]
ex=(0,0,0,0.1)
plt.pie(sizes,explode=ex,labels=names,autopct='%1.1f%%',shadow=False,startangle=90)
plt.title('模拟2019年电商市场占有率')
plt.savefig('D:/pyData/例9')
plt.show()
```
